```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
x = np.arange(0, 11) # range(10)
x
```

```python
y=2*x
y
```

```python
plt.figure()
plt.plot(x,y)
plt.title("my first plot")
plt.legend()
plt.xlabel("x dimension")
plt.ylabel("y dimension")
```

```python
#create a canvas with 2 rows and 2 columns
#(canvas = containing all the graphs together)

# first axis
plt.figure(figsize=(10,8))
plt.suptitle("this is my first canvas plotting")
plt.subplot(2,2,1)
plt.plot(x,y)
y=(x**7)-56

# second axis
plt.subplot(2,2,2)
y1=-x
plt.plot(x,y)
plt.xlabel("axes")

# third axis
plt.subplot(2,2,3)
y2=(x*3)+45
plt.plot(x,y2)
plt.plot(x,-x,color="red")

# fourth axis
plt.subplot(2,2,4)
y3=x**4
plt.plot(x,y3)
plt.plot(x,-x,color="#ff33cc")
#you can also pick the color from google and insert the color code e.g. #ff33cc for pink
```

```python
#placing more data in the same plot (1,1,1) --> 1 plot

plt.figure(figsize=(10,5))

plt.subplot(1,1,1)

y=x**4
plt.plot(x,y)
plt.plot(x,color="red")
plt.plot(x**3,color="#ff33cc")
plt.plot(x,color="#ff54cc")
```

```python
# create first figure and ax

y1 = x*2
y2 = x
y3 = x/2
y4 = x*3

# create the figure and axes
fig2,ax = plt.subplots(2,2,figsize = (10,6)) # Here the figsize controls the full canvas size, not the individual figure size

# add an overall title for the figure
fig2.suptitle('My second Matplotlib plot(s)', fontsize=18)



# we add a title to the figure or the ax
ax.set_title('My first Matplotlib plot', fontsize=18)
# we plot the different lines adding them to the axis
ax[0,0].plot(x,y1, label='y = 2x')
ax[1,1].plot(x,y2, c='red', label='y = x')
ax[0,0].plot(x,y3, c='green', label='y = x/2')
ax[0,0].plot(x,y4, c='orange', label='y = 3x')

# we add the legend
ax.legend()
```
