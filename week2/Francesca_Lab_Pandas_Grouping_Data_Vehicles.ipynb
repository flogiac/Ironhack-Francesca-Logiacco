# Pandas: grouping

```python
import pandas as pd
import numpy as np
```

```python
cars = pd.read_csv("data/vehicles.csv")
```

```python
# it return the first 5 rows of the data frame

cars.head()
```

### How many Car models? 

```python
### grouping by Models to see how many groups of car models we have


cars_model = cars.groupby(['Model']).count()
print(len(cars_model.index))
```

```python
# we have 3608 car models
```

### group by the data by the Make  using count function

```python
car_make = cars.groupby(['Make']).count()
car_make
```

```python
# numebr of cars per model

most_cars = car_make.groupby(['Make'])['Model'].max()
most_cars
```

What brand has the most cars?

```python
# numebr of maxmum number of cars
### question: there is a way to import here the name of the Maker directly???

most_cars.max()
```

```python
# by sorting we can get the name of the Maker with higher number of Models.

car_make.sort_values(by=['Model'], ascending = False)
```

### show the average CO2_Emission_Grams/Km  by Brand

```python
# changing column name Grams/Mile to Grams/Km

cars.rename(columns = {"CO2 Emission Grams/Mile" : "CO2 Emission Grams Km"}, inplace = True)

cars
```

```python
co_emission = cars.groupby(['Make'])['CO2 Emission Grams Km'].mean()
co_emission
```

### show the average CO2_Emission_Grams/Km  by Brand ... sorted

```python
co_emission.sort_values(ascending = False)
```

### Use `pd.cut` or `pd.qcut` to create 4 groups (bins) of cars, by Year. We want to explore how cars have evolved decade by decade.

```python
cars['Year'].describe()
```

```python
car_groups = pd.cut(cars['Year'], bins=4, labels=["<1985", "1995", "2005", ">2015"])
car_groups
```

## Did cars consume more gas in the eighties?

#### show the average City_Km/Liter by year_range

```python
# convert the column City MPG to City/Km

cars.rename(columns = {"City MPG" : "City Km/L"}, inplace = True)

cars
```

```python
cars.groupby(['Year','Make'])["City Km/L"].mean()
```

#### Which brands are more environment friendly?

```python
co2_emission = cars.groupby(['Make'])["CO2 Emission Grams Km"].mean()
co2_emission.sort_values(ascending=True)    
```

```python
# the Brand Fisker is more environment friendly since has less CO2 Emission.
```

### Do cars with automatic transmission consume more fuel than cars with manual transmission?

```python
Automatic_df = pd.DataFrame(cars.loc[cars['Transmission'].str.contains("Automatic")])
print(Automatic_df)
```

```python
Automatic_df = Automatic_df["Fuel Barrels/Year"].mean()
print(Automatic_df)
```

```python
Manual_df = pd.DataFrame(cars.loc[cars['Transmission'].str.contains("Manual")])
print(Manual_df)
```

```python
Manual_df = Manual_df["Fuel Barrels/Year"].mean()
print(Manual_df)
```

```python
Automatic_df > Manual_df
```
